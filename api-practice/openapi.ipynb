```python
#데이터 경로 response > body > items > item 에서 데이터 꺼내옴(데이터 타고 들어감)
# pprint(response_python["response"]["body"]["items"]["item"])

#item 데이터 추출출
item = response_python["response"]["body"]["items"]["item"]
pprint(item)
```

```python
#온도(기온 / T1H)를 나타내는 데이터 모두 출력
for i in item:
    #데이터 중 category 추출
    category = i["category"]
    #category가 T1H이면 해당날씨 데이터 출력
    if category=="T1H":
        print(i)
```

```python
#습도(REH)를 나타내는 데이터 모두 출력
#데이터 중 category가 REH인 경우
for i in item:
    #데이터 중 category 추출
    category = i["category"]
    #category가 T1H이면 해당날씨 데이터 출력
    if category=="REH":
        print(i)
```
