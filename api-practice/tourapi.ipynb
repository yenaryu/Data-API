```python
# requests 라이브러리 불러오기 + pprint 모듈 import
import requests
from pprint import pprint

# open api key
serviceKey = "cUXcvgk5JU5s8EWoLWYB28L5mjyVbhLTD8FhyY7JxmBYxqNHgJMh3tDNnzOaMG5Umh2QVsEFm1T4jmTGz9rhPg=="

# api url = base_url + endpoint
base_url = "http://apis.data.go.kr/B551011/KorService1/"
end_point = "locationBasedList1"  # 위치기반 관광정보 조회 API

# 요청 url 조합
request_url = base_url + end_point
print("request url:", request_url)

# 요청 변수 설정정
numOfRows = 10  # 한페이지 결과 수
pageNo = 1  # 페이지번호
MobileOS = "AND"  # 호출OS
MobileApp = "kakaotalk"  # 호출 어플이름
mapX = 126.981611  # GPS X좌표 (서울경도)
mapY = 37.568477  # GPS Y좌표 (서울위도)
radius = 1000  # 거리반경
_type = "JSON"  # 응답 데이터의 형식

# 요청 변수 저장
params = {
    "serviceKey": serviceKey,
    "numOfRows": numOfRows,
    "pageNo": pageNo,
    "MobileOS": MobileOS,
    "MobileApp": MobileApp,
    "mapX": mapX,
    "mapY": mapY,
    "radius": radius,
    "_type": _type,
}

# API 요청
response = requests.get(request_url, params=params)
response_python = response.json()

item = response_python["response"]["body"]["items"]["item"]
pprint(item)
```
