```python
#라이브러리 불러오기
import requests

#API요청 URL
url = "https://jsonplaceholder.typicode.com/posts"

#URL 접속(API요청)
response_data = requests.get(url)

print(response_data)
```

```python
#응답 데이터의 데이터형식을 json에서 python으로 변환
response_python = response_data.json()
print(response_python)
```

```python
from pprint import pprint
pprint(response_python)
```

```python
#라이브러리 불러오기
import requests

#API요청 URL
url = "https://jsonplaceholder.typicode.com/posts/1"

#URL 접속(API요청)
response_data = requests.get(url)
print(response_data)

#응답 데이터의 데이터형식을 json에서 python으로 변환
response_python = response_data.json()
print(response_python)

from pprint import pprint
pprint(response_python)
```
