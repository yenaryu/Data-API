```python
# 현재 상영 중인 영화 목록을 1 페이지부터 5 페이지까지 요청하고, 응답 데이터 리스트에 저장하고, 출력하기
import requests
from pprint import pprint

api_token = "eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiIxYzI5OTlkNGQyZmM3YmMyYmViMmQ5OGVkZjQ1Njk0MiIsIm5iZiI6MTczNDM5NTE4My43ODIsInN1YiI6IjY3NjBjNTJmNjExNjYxZTEwNDZiZTlmYiIsInNjb3BlcyI6WyJhcGlfcmVhZCJdLCJ2ZXJzaW9uIjoxfQ.k5lziHC6UucR1Skiyg0KergiWYlGTFgwShQZQZpnOLY"

base_url = "https://api.themoviedb.org/3/"
endpoint = "movie/now_playing"
request_url = base_url + endpoint

# 1페이지부터 5페이지까지 영화 정보를 저장할 리스트트
results_list = []

for page in range(1, 6):
    params = {"language": "ko-KR", "page": page}  # 페이지정보
    headers = {"accept": "application/json", "Authorization": f"Bearer {api_token}"}

    response_data = requests.get(request_url, params=params, headers=headers)

    response_python = response_data.json()

    # result의 데이터타입이 list
    results = response_python["results"]
    # 리스트.extend(입력리스트) : 입력리스트를 풀어서 리스트에 추가한다다
    results_list.extend(results)

print(len(results_list))
pprint(results_list)

# results에서 영화 정보 가져오기
# movies = response_python.get("results", [])

# # 첫 3개의 영화만 출력
# three_movies = movies[:3]

# for i in range(len(three_movies)):  # three_movie의 길이 range로 반복
#     movie = three_movies[i]  # 인덱스 i번째 요소인 영화를 가져옴
#     print(f"영화 {i + 1}:")  # 인덱스+1번째
#     print(movie)
```
