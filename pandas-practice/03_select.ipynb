```python
import pandas as pd
```

```python
#csv파일 불러오기
df = pd.read_csv("./Car.csv")
df.head(5) #상위 5개 출력
```

```python
#열 선택
df['name']
```

```python
#2개 이상의 열을 선택할 때는 열들을 리스트에 넣어서 선택
df[['name', 'fuel']]
```

```python
#행 슬라이싱
df[0:] #시작부터 끝까지 모두 슬라이싱 (python list slicing)
df[5:10] #인덱스 위치 기반으로 슬라이싱
```

```python
#.iloc[] 인덱싱 : `컬럼의 위치`를 기반으로 한 선택 방법
df.iloc[0] #첫번째 행 데이터 도출
df.iloc[0:10] #행 슬라이싱 (0~9행) 리스트의 슬라이싱처럼 종료 위치(10)이 포함되지 않음

#열 선택 (df.iloc[행,열])
#리스트[:] - 리스트의 모든 원소
df.iloc[:,0] #모든 행(:)에 대한 첫번째열(0)
```

```python
df.head()
# 모든 행을 선택하는데 첫번째 두번째 세번째 열을 선택하여 출력하라
df.iloc[:,0:3]

# 5:10행을 선택하는데 0:6열을 선택해서 출력하라
df.iloc[5:10,0:6]
```

```python
#.loc[] : 레이블 기반 선택. `행 이름`을 보고 선택해서, 컬럼만 보고도 알 수 있음. 가장 많이 사용
#.loc[행,열]

#모든 행을 선택
df.loc[:]

#행 이름이 "0"인 행을 선택
df.loc[0]

#행 이름이 "0", "1", "3"인 데이터 선택
df.loc[[0,1,3]]

#슬라이싱 (종료 레이블 포함)
df.loc[0:10]

#단일 열(name)선택
df.loc[:, "name"]

#복수열(name, seller_type)선택
df.loc[:, ["name", "seller_type"]]

#행과 열 선택 조합
df.loc[[0,3,5], ["name", "selling_price"]]
df.loc[0:10, ["name", "selling_price"]]
```

```python
#1. selling_price 열 기준으로 내림차순으로 정렬
df.sort_values(by = "selling_price", ascending=False)
```

```python
#2. 결과를 새로운 변수 (sorted_df)에 저장
sorted_df = df.sort_values(by = "selling_price", ascending=False)
sorted_df
```

```python
#3.sorted_df에 대해 인덱스 기반 선택
sorted_df.iloc[0]
```

```python
#4. sorted_df에 대해 레이블 기반 선택
sorted_df.loc[0]
```

```python
#조건식
#열을 선택해서 해당 열의 개별데이터와 값을 비교하는 식
#판매가격(selling_price) 100000보다 작은

#1. 열 선택
#2. 조건 값을 작성
#3. 비교연산자 작성
condition = df['selling_price'] < 100000

#4. loc를 사용해서 데이터를 선택(필터링)
df.loc[condition] #True인 행만 반환해서 보여줌줌
```

```python
#주행거리(km_driven)이 100000보다 큰 데이터를 필터링
driven_data = df['km_driven'] > 100000
df.loc[driven_data] 
```

```python
#연료(fuel)가 "Diesel"인 데이터를 필터링
fuel_data = df['fuel'] == 'Diesel'
df.loc[fuel_data] 
```

```python
#판매유형(seller_type)이 "Dealer"가 아닌 데이터를 필터링

seller_data = df['seller_type'] != 'Dealer'
df.loc[seller_data] 
```

```python
#주행거리(km_driven)이 100000보다 작은
condition01 = df['km_driven'] > 100000
#판매가격(selling_price)이 100000보다 작은
condition02 = df['selling_price'] < 100000

#두 조건을 모두 만족하는 데이터를 선택
df.loc[condition01 & condition02] 
```

```python
#주행거리(km_driven)이 100000보다 작은
condition01 = df['km_driven'] > 100000
#판매가격(selling_price)이 100000보다 작은
condition02 = df['selling_price'] < 100000
#운전방식(transmission)이 Automatic인 차량
condition03 = df['transmission'] == "Automatic"

#두 조건을 모두 만족하는 데이터를 선택택
df.loc[condition01 & condition02 & condition03] 
```

```python
#fuel가 Electric 이거나 transmission이 Automatic인 차량을 선택

condition01 = df['fuel'] == "Electric"
condition02 = df['transmission'] == "Automatic"
df.loc[condition01 | condition02] 
```

```python
#year가 2018보다 크면서
#fuel가 Petrol 또는 Diesel인 데이터 선택

condition01 = df['year'] > 2018
condition02 = (df['fuel'] == "Petrol") | (df['fuel'] == "Diesel")

#두 조건을 모두 만족하는 데이터를 선택
df.loc[condition01 & condition02]
```

```python
#연료(fuel)가 Petrol, Diesel, LPG인 데이터를 탐색
condition = df["fuel"].isin(['Petrol', 'Diesel', 'LPG'])

#연료(fuel)가 Petrol, Diesel, LPG인 데이터를 선택
df.loc[condition]

#연료(fuel)가 Petrol, Diesel, LPG이 아닌 데이터를 선택
df.loc[~condition]
```

```python
#year가 2020, 2018, 2016, 2014인 데이터를 선택

condition = df["year"].isin([2020, 2018, 2016, 2014])
df.loc[condition]
```
