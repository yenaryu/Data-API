```python
#str접근자
#문자열 메서드를 활용하기 위한 도구
df["이름"].str
```

```python
#알파벳 대/소문자 변환
#.str.lower() : 소문자 / .str.upper() : 대문자자

#이메일 데이터를 소문자로 변환
df["이메일"].str.lower()
```

```python
#데이터 프레임을 직접적으로 수정하지 않는다. 생성한 데이터를 다시 열(df['이메일'])에 저장해야한다.
df["이메일"] = df["이메일"].str.lower()
df
```

```python
#특정 문자열이 포함되어있는지 확인
#str.contains("문자열")
#괄호안에 작성한 문자열을 탐색. 이메일에서 ".error" 포함되어있는지 여부 확인
condition = df["이메일"].str.contains(".error")
condition
```

```python
#데이터 필터링
df.loc[condition]
```

```python
#문자열을 특정 문자열로 대체
#str.replace("기존 문자열", "대체 문자열")
#이메일의 ".error"을 ".com"으로 대체
df["이메일"] = df["이메일"].str.replace(".error", ".com")
df
```

```python
#특정 문자열로 시작 / 끝 확인
#str.startswith("문자열") / str.endswith("문자열")

#성별 "남"으로 시작하는 데이터 확인
condition = df["성별"].str.startswith("남")

#데이터 필터링
df.loc[condition]
```

```python
#문자열 분할
#str.split("특정 문자열") : 특정 문자열을 기준으로 문자열을 분할. 문자열을 구분하는 역할(구분자)

#이메일을 "@"를 기준으로 분할
df["이메일"].str.split("@")
```

```python
#이메일을 "@"를 기준으로 분할 후, 첫번째 값 추출하여 "아이디"에 저장
df["아이디"] = df["이메일"].str.split("@").str[0]
df
```

```python
#이메일을 "@"를 기준으로 분할 후, 두번째 값 추출하여 "도메인"에 저장
df["도메인"] = df["이메일"].str.split("@").str[1]
df
```

```python
#공백 제거
#str.strip() : 양쪽 공백 제거
#str.lstrip() : 왼쪽 공백 제거 (시작공백)
#str.rstrip() : 오른쪽 공백 제거 (끝공백)

#이름에서 양쪽 공백 제거
df["이름"] = df["이름"].str.strip()
df
```

```python
#문자열의 길이 계산
#str.len()

#이름의 길이(문자) 계산
df["이름"].str.len()
```

```python
# 2. 전화번호에서 지역 번호를 추출하고, 열 “지역 번호”를 생성하고, 데이터를 저장한다.
df["지역 번호"] = df["전화번호"].str.split("-").str[0]
df
```

```python
# 2-1. 휴대폰 번호 앞자리가 아닌, 진짜 지역번호만 추출
df["지역번호2"] = df.loc[~df["전화번호"].str.contains("010"),"전화번호"].str.split("-").str[0]
df["지역번호2"] = df["지역번호2"].fillna("") #NaN 빈값으로
df
```

```python
# 3. 이메일에서 아이디를 추출하고, 열 “아이디”를 생성하고, 데이터를 저장한다.
df["아이디"] = df["이메일"].str.split("@").str[0]
df
```

```python
# 4. 주소에서 도시명을 추출하고, 열 “지역”을 생성하고, 데이터를 저장한다.
df["지역"] = df["주소"].str.split("시").str[0]
df
```

```python
# 5. 지역에서 “시”를 모두 제거한다.
df["지역"] = df["지역"].str.replace("시", "")
df
```
