```python
#조건식 사용은 불가능하다
#condition = df.groupby("부서")["연봉"].mean() > 6000

#조건식 대신
#Dataframe.groupby("열").filter(함수)

#부서의 인원수가 2명보다 많은 부서를 필터링
#부서 당 인원 수 출력

#1. 그룹화
grouped = df.groupby("부서")

# 조건함수
# 매개변수 : 그룹화한 데이터의 개별행
def greater_count_two(group):
    count = group["이름"].count()
    #count가 2보다 크면 결과값이 True
    if count > 2:
        return True

    #count가 2보다 작으면 결과값이 False
    if count <= 2:
        return False

def only_true(group):
    return True

#2. 필터링
grouped.filter(greater_count_two)
```

```python
#1. 그룹화
grouped = df.groupby("부서")

#2. 조건함수
#매개변수 : 부서를 기준으로 그룹화된 행 데이터
def greater_salary(group):
    mean_salary = group["연봉"].mean()

    if mean_salary >= 6000:
        return True
    
    else:
        return False

#3. 필터링 filter() 적용
grouped.filter(greater_salary)
```

```python
#평균 연봉이 6000이상인 부서만 필터링
#1. 그룹화
group_rows = df.groupby("부서")

#2.조건함수
def greater_salary(group_row):
    mean_salary = group_row["연봉"].mean()

    return  mean_salary >= 6000

#3. 필터링
grouped.filter(greater_salary)
```

```python
#연봉 열에 평균, 중위값
#적용하고 싶은 집계함수를 리스트 문자열로 입력
#mean, median
df["연봉"].agg(["mean", "median"])

#나이열의 최소, 최대, 평균
df["나이"].agg(["min", "max", "mean"])
```

```python
#여러개의 열에 여러개의 집계함수
"""
딕셔너리(key-value)를 활용해서 열과 열에 적용할 집계함수 작성
key : 열
value : 집계함수

Dataframe.agg(
{
"열1" : "min",
"열2" : ["min","max"]
})
"""
#이름의 개수
#연봉의 평균, 중위값
#나이의 최소, 최대, 평균

df.agg(
    {
        "이름" : "count",
        "연봉" : ["mean","median"],
        "나이" : ["min", "max", "mean"]
    }
)
```

```python
#그룹화된 데이터에도 다중 집계 함수 agg()
grouped = df.groupby("지역")
result = grouped.agg(
    {
        "이름" : "count",
        "연봉" : ["mean","median"],
        "나이" : ["min", "max", "mean"]
    }
)

#다중 집계함수의 결과 : 컬럼들이 계층구조를 가진다
#만약에 연봉 - mean 선택하고 싶다면?
result[("연봉","mean")]

#나이 - min, max 선택택
result.loc[:, ("나이",["min","max"])]
```
