```python
#columns 속성 : 열의 이름 정보를 저장함
df.columns

#열 이름을 수정할 수 있다
#다만 리스트의 길이와 기존 열 개수가 동일해야한다.
# df.columns = ["A", "B", "C", "D"]
# print(df.columns)
```

```python
#sort_values(by=[]) : 열의 값을 기준으로 정렬
#by : 어떤 열을 기준으로 정렬?
#ascending : 오름차순
#주의사항 : 정렬한 결과가 원본 데이터 프레임에 적용되지 않음. 기존 데이터프레임은 변화X

#"volatile acidity"를 기준으로 오름차순으로 정렬
df.sort_values(by = "volatile acidity", ascending=True)
```

```python
#"fixed acidity"를 기준으로 내림차순으로 정렬
df.sort_values(by = "fixed acidity", ascending=False)
```

```python
#quality, alcohol기준 오름차순 정렬
sorted_df = df.sort_values(by=["quality","alcohol"])
sorted_df
```

```python
#quality, alcohol, residual sugar 기준 오름차순 정렬
sorted_df = df.sort_values(by=["quality","alcohol","residual sugar"])
sorted_df
```

```python
#quality 기준 내림차순 / alcohol, residual sugar 기준 오름차순 정렬
sorted_df = df.sort_values(by=["quality","alcohol","residual sugar"],ascending=[False, True, True])
sorted_df
```

```python
#sort_index() : 행의 이름(index)를 기준으로 정렬

#행의 이름은 index
#열의 이름은 column
# 인덱스를 기준으로 오름차순 정렬
sorted_df = df.sort_index()
sorted_df
```

```python
# 인덱스를 기준으로 내림차순 정렬
sorted_df = df.sort_index(ascending=False)
sorted_df
```
