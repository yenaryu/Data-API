```python
import pandas as pd
```

```python
#csv파일 불러오기
df = pd.read_csv("./titanic.csv")
```

```python
# 1. 첫 번째 행만 선택하여 출력한다.
df.head(1)
```

```python
# 2. 열 age 만 선택하여 출력한다.
df['age']
```

```python
# 3. 열 age fare 를 선택하여 출력한다.
df[['age', 'fare']]
```

```python
# 4. 0행 부터 5행까지 출력한다.
df[0:6]
```

```python
# 5. 열 pclass fare 을 선택하여 모든 행을 출력한다.
condition=df[['pclass', 'fare']]
condition.loc[:]
```

```python
# 6. 열 age 가 1 미만인 행만 출력한다.
condition=df['age']<1
df.loc[condition]
```

```python
# 7. 열 `age` 가 30세 이상인 행만 출력한다.
condition=df['age']>30
df.loc[condition]
```

```python
# 8. 열 `age` 가 30세 이상인 행의 열 `survived` 만 출력한다.
condition=df['age']>30
df.loc[condition, 'survived']
```

```python
# 9. 열 `survived` 가 1인 모든 행을 출력한다.
condition=df['survived'] == 1
df.loc[condition]
```

```python
# 10. 열 `pclass` 가 1인 모든 행을 출력한다.
condition=df['pclass'] == 1
df.loc[condition]
```

```python
# 11. 열 `pclass` 가 1이고, 열 `survived` 가 1인 모든 행을 출력한다.
condition1=df['pclass'] == 1
condition2=df['survived'] == 1
df.loc[condition1 & condition2]
```

```python
# 12. 열 `pclass` 가 2 또는 3인 모든 행을 출력한다.
condition = df["pclass"].isin([2,3])
df.loc[condition]
```

```python
# 13. 열 `pclass` 가 2 또는 3이고, `survived` 가 1인 모든 행을 출력한다.
condition = (df["pclass"].isin([2,3])) & (df["survived"] == 1)
df.loc[condition]
```
