```python
import pandas as pd

#(1) 2개의 타이타닉 데이터 불러오기
df2 = pd.read_csv("./titanic_2.csv")
df3 = pd.read_csv("./titanic_3.csv")
```

```python
#(2) 데이터 합치기
df = pd.concat([df2, df3], ignore_index=True)
df
```

```python
# (3) 중복 삭제하기

# 1. 데이터의 중복 데이터를 출력한다.
df.duplicated(keep=False)
```

```python
# (3) 중복 삭제하기

# 2. 중복 데이터가 있다면 중복 데이터를 삭제한다. 
# 이때 새로운 변수 `unique_df` 를 생성하고, 중복을 제거한 데이터프레임을 저장
unique_df = df.drop_duplicates()
unique_df
```

```python
# (4) 열 삭제하기
# 열을 삭제한다 : `alive` `embarked` `class` `deck` `who`
unique_df.drop(columns=['alive','embarked','class','deck','who'])
```

```python
# ### (5) 값 수정하기
# 1. 열 `fare` 데이터 중 0인 행의 `fare` 값을 10으로 수정한다.
unique_df.loc[unique_df["fare"] == 0, "fare"] = 10
unique_df
```

```python
# `fare` 을 기준으로 오름차순으로 정렬한 값을 출력해서 0인 행이 존재하는지 확인한다.
sorted_df = unique_df.sort_values(by="fare")
sorted_df
```

```python
# 1. 열 `age` 데이터 중 1보다 작은 행의 `age` 값을 1로 수정한다.
unique_df.loc[unique_df["age"] < 1, "age"] = 1
unique_df
```

```python
# `age` 를 기준으로 오름차순으로 정렬한 값을 출력해서 1보다 작은 값이 존재하는지 확인한다.
sorted_df = unique_df.sort_values(by="age")
sorted_df
```
