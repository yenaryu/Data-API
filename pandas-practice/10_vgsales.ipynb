```python
import pandas as pd

df = pd.read_csv("./vgsales.csv")
```

```python
#3. 장르별 세계 판매량 합계를 출력한다.
grouped = df.groupby("genre")
sum_grouped = grouped['total_sales'].sum().sort_values(ascending=False)
sum_grouped
```

```python
# 4. 플랫폼별 세계 판매량 평균을 출력한다.
grouped = df.groupby("console")
mean_grouped = grouped['total_sales'].mean().sort_values(ascending=False)
mean_grouped
```

```python
# 5. 개발사별 게임 출시 개수를 출력한다.
grouped = df.groupby("developer")
count_grouped = grouped['title'].count().sort_values(ascending=False)
count_grouped
```

```python
# 6. 플랫폼별 세계 판매량, 북미 판매량, 일본 판매량, 유럽 및 지역 아프리카 판매량, 기타 지역 판매량의 합계와 평균을 출력한다.

grouped = df.groupby("console")[["total_sales","na_sales","jp_sales","pal_sales","other_sales"]].agg(["sum", "mean"]).sort_values(by=("jp_sales", "sum"), ascending=False)
grouped
```

```python
# 7. 평론가 점수의 평균이 9.5가 넘는 게임 개발사를 선택해서 출력한다.
grouped = df.groupby("developer")

# mean_grouped = grouped['critic_score'].mean()
# mean_grouped = mean_grouped[mean_grouped > 9.5]
# mean_grouped


def greater_score(group):
    return group['critic_score'].mean() > 9.5

grouped.filter(greater_score)
```
