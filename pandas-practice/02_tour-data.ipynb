```python
## 위치기반 관광정보 조회 API 데이터

#1. OPEN API 데이터의 정보 3개를 확인하는 코드 작성

# requests 라이브러리 불러오기 + pprint 모듈 import
import requests
import pandas as pd
from pprint import pprint

serviceKey = "cUXcvgk5JU5s8EWoLWYB28L5mjyVbhLTD8FhyY7JxmBYxqNHgJMh3tDNnzOaMG5Umh2QVsEFm1T4jmTGz9rhPg=="
base_url = "http://apis.data.go.kr/B551011/KorService1/"
end_point = "locationBasedList1"
request_url = base_url + end_point

# 요청 변수 설정
numOfRows = 10  # 한페이지 결과 수
pageNo = 1  # 페이지번호
MobileOS = "AND"  # 호출OS
MobileApp = "kakaotalk"  # 호출 어플이름
mapX = 126.981611  # GPS X좌표 (서울경도)
mapY = 37.568477  # GPS Y좌표 (서울위도)
radius = 1000  # 거리반경
_type = "JSON"  # 응답 데이터의 형식

# 요청 변수 저장
params = {
    "serviceKey": serviceKey,
    "numOfRows": numOfRows,
    "pageNo": pageNo,
    "MobileOS": MobileOS,
    "MobileApp": MobileApp,
    "mapX": mapX,
    "mapY": mapY,
    "radius": radius,
    "_type": _type,
}

# API 요청
response = requests.get(request_url, params=params).json()
# pprint(response)
items = response["response"]["body"]["items"]["item"]

# 정보를 가지고, data변수 생성
# 필요 데이터 : title, addr1, mapx
data = [
    {
      "title": item["title"],
        "addr1": item["addr1"],
        "mapx": item["mapx"],
    }
     for item in items
]
# pprint(data)

df = pd.DataFrame(data)
pprint(df)

#tour_data.csv 저장
df.to_csv("./tour_data.csv")
```
