**Scatter Plot 산점도**

```python
#tips 데이터를 활용한 산점도

#모듈 불러오기
import seaborn as sns
import matplotlib.pyplot as plt

#데이터셋 불러오기
tips_df = sns.load_dataset("tips")
tips_df.info()
```

```python
tips_df.head()
```

```python
#산점도(Scatter plot)
#2개의 연속형 데이터를 기반으로

#total_bill 에 따른 tip을 산점도 시각화
#sns.scatterplot(data, x, y, hue)
sns.scatterplot(data=tips_df,x="total_bill",y="tip")

#그래프를 출력
plt.show()
```

```python
#점(포인트)를 구분하는 범주형 데이터
#hue 매개변수 : 포인트를 구분하는 범주형 데이터

sns.scatterplot(data=tips_df,x="total_bill",y="tip",hue="size")
plt.show()
```

**Line Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

#sns.lineplot(data, x, y)
df = sns.load_dataset("flights")

# 기본 라인 플롯
sns.lineplot(data=df, x="year", y="passengers")
plt.title("Basic Line Plot")
plt.show()
```

**FacetGrid**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seaborn의 tips 데이터셋 로드
tips = sns.load_dataset("tips")

# FacetGrid 생성
# 열 기준 : 요일(day)

#g = sns.FacetGrid(dataframe, row, col)
#g.map(plot)
g = sns.FacetGrid(tips, col="day")


# 서브플롯의 종류 : 히스토그램(histogram)
# 기준 값 : 총 청구 금액
g.map(plt.hist, "total_bill")

plt.show()
```

```python
import seaborn as sns

titanic = sns.load_dataset("titanic")

print(titanic.head())
```

```python
# 2. `age` 와 `fare` 의 관계를 Scatter Plot으로 시각화한다.
import matplotlib.pyplot as plt
sns.scatterplot(data=titanic, x="age", y="fare")
plt.show()
```

```python
# 3. `age` 와 `fare` 의 관계를 `survived` 로 구분해서 Scatter Plot으로 시각화한다.
import matplotlib.pyplot as plt
sns.scatterplot(data=titanic,x="age",y="fare",hue="survived")
plt.show()
```

```python
import seaborn as sns

diamonds = sns.load_dataset("diamonds")

print(diamonds.head())
```

```python
# 2. `price` 와 `carat` 의 관계를 Scatter Plot으로 시각화한다.
import matplotlib.pyplot as plt
sns.scatterplot(data=diamonds, x="price", y="carat")
plt.show()
```

```python
# 3. `price` 와 `carat` 의 관계를 `color` 로 구분해서 Scatter Plot으로 시각화한다.
import matplotlib.pyplot as plt
sns.scatterplot(data=diamonds, x="price", y="carat",hue="color")
plt.show()
```

```python
# 4. `price` 와 `carat` 의  관계를 `cut` 로 구분해서 Scatter Plot으로 시각화한다.
import matplotlib.pyplot as plt
sns.scatterplot(data=diamonds, x="price", y="carat",hue="cut")
plt.show()
```

```python
# 5. `price` 와 `carat` 의 관계를 `cut` 품질에 따라 구분해서 다중 시각화한다.
import matplotlib.pyplot as plt
sns.lineplot(data=diamonds, x="price", y="carat")
plt.show()
```
