#### 관계형 데이터 시각화

```python
import seaborn as sns
import matplotlib.pyplot as plt

#seaborn 실습 데이터 tips
tips_df = sns.load_dataset("tips")
tips_df.head()
```

```python
tips_df["time"].unique() #time열의 범주형 데이터 확인
```

```python
#시각화 요소
#데이터, x축, y축, 구분데이터(hue)
sns.scatterplot(data=tips_df,x="total_bill",y="tip",hue="time")
plt.show()
```

```python
flight_df = sns.load_dataset("flights")
flight_df.head()
```

```python
#line plot(data,x,y)
sns.lineplot(data=flight_df,x="year",y="passengers")
plt.show()
```

```python
#월별 승객수
sns.lineplot(data=flight_df,x="month",y="passengers")
plt.show()
```

```python
#FacetGrid
#Grid : 격자 형태로 여러개 그래프(서브플롯) 시각화해서 관계를 확인한다
tips_df = sns.load_dataset("tips")
```

```python
grid = sns.FacetGrid(data = tips_df,
                     col="smoker", #구분을 위한 범주형데이터(day, smoker, time, ...)
                     )

#그리드의 그래프 시각화
grid.map(plt.hist, "total_bill")
plt.show()
```
