#### 분포형 데이터 시각화

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#펭귄데이터
penguin_df = sns.load_dataset("penguins")
penguin_df.head()
```

```python
penguin_df.shape #데이터 프레임의 형상(행과 열의 수)
```

```python
#펭귄 부리길이 분포
sns.histplot(data=penguin_df, x="bill_length_mm")
plt.show()
```

```python
#펭귄종에 따른 부리의 길이 분포포
sns.histplot(
    data=penguin_df,
    x="bill_length_mm",
    bins=10, #히스토그램 bin(막대의 개수) 조절
    hue="species",
    multiple="layer" #multiple : layer(기본값) 겹쳐서 표현현 / stack 쌓아서 구분해서 표현 / dodge 옆으로 나란히 구분해서 표현 / fill 각 범주의 막대 높이를 비율로 표현현
)
plt.show()
```

```python
#펭귄종에 따른 날개길이 분포 KDE plot
sns.kdeplot(
    data=penguin_df,
    x="flipper_length_mm",
    hue="species",
    fill=True,
    bw_adjust=10, #대역폭 / 값이 커지면 곡선이 부드러워지고, 작아지면 곡선이 날카로워짐 (기본값 1)
)
plt.show()
```

```python
sns.rugplot(data=penguin_df, x="body_mass_g")
plt.show()
```
