#### 범주형 데이터 시각화

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris_df = sns.load_dataset("iris")
iris_df.head()
```

```python


sns.barplot(
    data=iris_df, 
    x="species", #품종
    y="sepal_length", #꽃받침 길이이
)
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("./car_evaluation.csv")
```

```python
df.head()
# df.info()
```

```python
sns.countplot(
    data = df,
    y="class",
    order=["vgood","good","acc","unacc"], #출력순서
    hue="price" #
)
plt.show()
```

```python
sns.countplot(
    data = df,
    y="safety",
    hue="class"
)
plt.show()
```
