#### 기타 시각화

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_df = sns.load_dataset("titanic")
titanic_df.info()
```

**Heatmap**

```python
#수치형 데이터만 선택
numeric_df = titanic_df.select_dtypes(include=['float64', 'int64'])

#상관계수 행렬 계산
matrix = numeric_df.corr()

#heatmap
sns.heatmap(
    matrix,
    annot=True,
    cmap="plasma"
)
plt.show()
```

**Pie Chart**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./movie_ratings.csv")
df.head()

#범주형 데이터의 비율(빈도수)
#value_counts()

#pie chart의 값(크기)
values = df["Genre"].value_counts()

# pie label
labels = df["Genre"].unique()

plt.pie(x=values, labels=labels, autopct="%1.1f%%")
plt.show()
```
